```python
#NN-align encoding of IC50 values
# The peptide binding affinity IC50 values were encoded to the neural network as log-transformed values, 
#using the relation 1-log (aff)/log(50000), 
#where aff is the measured binding affinity (IC50) in nM units [29].
nnencb = 1+(pic50b/np.log10(50000))
nnenca = 1+(pic50a/np.log10(50000))
plt.figure(figsize=(15,7))
plt.hist(nnencb, bins = 10, label = 'below 500 nM')
plt.hist(nnenca, bins = 10, label = 'above 500 nM')
plt.xlabel('1-{log10(ic50)/log10(50000)}')
plt.ylabel('count')
plt.legend()
plt.show()
```

```python
lengths = []
aa_counts = []
aa_dict = {
'A':[],
'R':[],
'N':[],
'D':[],
'C':[],
'E':[],
'Q':[],
'G':[],
'H':[],
'I':[],
'L':[],
'K':[],
'M':[],
'F':[],
'P':[],
'S':[],
'T':[],
'W':[],
'Y':[],
'V':[],
'X':[], #UNKNOWN
}
for index, row in bench.iterrows():
    peptide = row['peptide_sequence']
    counted = Counter(peptide)
    for key in aa_dict:
        if key in counted.keys():
            aa_dict[key].append(counted[key])
        else:
            aa_dict[key].append(0)
    lengths.append(len(peptide))
```

```python
#Look into modifications on peptides
#Encode amino acids
peptide_enc = []
mod_index = []
ic50_df['aa_enc'] = ''
for index, row in ic50_df.iterrows():
    seq = row['peptide_sequence']
    if '+' in seq:
        print(index, seq)
        mod_index.append(index)
        continue
    else:
        enc = encode_aa(seq)
        peptide_enc.append(np.asarray(enc))
        ic50_df['aa_enc'].iloc[index] = enc
```

```python
#Has to start at 0, since Keras apparently uses the index for lookup
AMINO_ACIDS = { 
'A':0,
'R':1,
'N':2,
'D':3,
'C':4,
'E':5,
'Q':6,
'G':7,
'H':8,
'I':9,
'L':10,
'K':11,
'M':12,
'F':13,
'P':14,
'S':15,
'T':16,
'W':17,
'Y':18,
'V':19,
'X':20, #UNKNOWN
'-':21
}
def encode_aa(sequence):
    enc = []
    for a in sequence:
        try:
            enc.append(AMINO_ACIDS[a])
        except:
            pdb.set_trace()
    return enc
```
