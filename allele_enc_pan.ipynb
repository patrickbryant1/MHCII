```python
def encode_alleles(alleles, ACHAINS_ENC, BCHAINS_ENC):
    '''Encode the alleles by getting matching sequences
    '''
    CHAIN1 = []
    CHAIN2 = []
    for a in alleles:
        if 'HLA' in a:
            a = a.split('-')
            CHAIN1.append(BCHAINS_ENC[a[1]])
            CHAIN2.append(BCHAINS_ENC[a[2]])
        if 'H-2-I' in a:
            CHAIN1.append(ACHAINS_ENC['H2M'])
            CHAIN2.append(BCHAINS_ENC[a])
        else:
            if 'DR' in a:
                CHAIN1.append(ACHAINS_ENC['DRB1_1402'])
            if 'DP' in a:
                CHAIN1.append(ACHAINS_ENC['HLADP2'])
            if 'DQ' in a:
                CHAIN1.append(ACHAINS_ENC['HLADQ2.5'])
            CHAIN2.append(BCHAINS_ENC[a])
    return np.asarray(CHAIN1), np.asarray(CHAIN2)
```
